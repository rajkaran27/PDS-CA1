### What is the difference in the prices of flat types?

```python
# getting the different flat types
flat_types = np.unique(data_resale_flat_prices['flat_type'])

# getting the average resale prices for each flat type
# creating an numpy array with the length of flat types
resale_prices = np.zeros(len(flat_types))
# finding average resale price of each flat type and storing them into resale_prices array
for i in range(len(flat_types)):
    resale_prices[i] = np.mean(data_resale_flat_prices['resale_price'][data_resale_flat_prices['flat_type'] == flat_types[i]])

# Plotting the graph
plt.figure(figsize=(20, 10))
plt.bar(flat_types, resale_prices)
plt.xlabel('Flat Type',fontsize=15)
plt.ylabel('Average Resale Price',fontsize=15)
plt.title('Average Resale Prices Across Different Flat Types',fontsize=20)
plt.grid(color="gray",alpha=0.7)
plt.xticks(rotation=45)
plt.show()
```

This is a bar chart showing the different Flat types and their prices. There are 7 flat types in this bar chart. They are 1 Room, 2 Room, 3 Room, 4 Room, 5 Room, Executie and Multi Generation.

The cheapest flat type is the 1 Room flat, which costs close to 200000SGD. The most expensive flat type is the Multi-Generation Flat, costing more than 800000SGD. As the number of rooms increases, the price of the flat increases too. The Executive, 5 Room and 4 Room flat have the same number of bedrooms, with 3 bedrooms. However, the executive flat has more floor area compared to the 5 room and 4 room flats. The Multi-Generational flat has the same floor area as a 5 room flat. However, the Multi-Generational flat has two bedrooms with attached bathrooms, and two other bedrooms. This is why the Multi-Generational flat costs the most, at around 800000SGD.

As my budget is 500000SGD, I will choose the 4 room flat.


### Which flat type had the most sales?

```python
count_1_room = np.sum(data_resale_flat_prices['flat_type']=='1 ROOM')
count_2_room = np.sum(data_resale_flat_prices['flat_type']=='2 ROOM')
count_3_room = np.sum(data_resale_flat_prices['flat_type']=='3 ROOM')
count_4_room = np.sum(data_resale_flat_prices['flat_type']=='4 ROOM')
count_5_room = np.sum(data_resale_flat_prices['flat_type']=='5 ROOM')
count_executive = np.sum(data_resale_flat_prices['flat_type']=='EXECUTIVE')
count_mg = np.sum(data_resale_flat_prices['flat_type']=='MULTI-GENERATION')

count_12_and_mg = count_1_room + count_mg +count_2_room

sizes = [count_3_room, count_4_room, count_5_room, count_executive, count_12_and_mg]


labels = np.unique(data_resale_flat_prices['flat_type'])

new_labels = []
for label in labels:
    if label == '1 ROOM':
        np.delete(label, np.where(label == '1 ROOM'))
    elif label == '2 ROOM':
        np.delete(label, np.where(label == '2 ROOM'))
    elif label == 'MULTI-GENERATION':
        continue
    else:
        new_labels.append(label)

new_labels.append('1 ROOM + 2 ROOM + MULTI-GENERATION')

explode =(0,0.1,0,0,0)


fig1,ax1 = plt.subplots(figsize=(8,8),facecolor='white')
ax1.pie(sizes,explode=explode,shadow=True, autopct='%1.1f%%',textprops={'fontsize': 15})
plt.legend(new_labels,title="Flat Types",loc="center left" ,bbox_to_anchor=(1, 0.5))
plt.title("Distribution of Flat Types Sold",fontsize=20)
plt.show()
```

### What is the relationship between the lease commence date and the resale prices of 4 room flats?

```python
room_4 = data_resale_flat_prices[data_resale_flat_prices['flat_type']=='4 ROOM']

lease_commence_date = np.unique(room_4['lease_commence_date'])
avg_price = []
for date in lease_commence_date:
    avg_price.append(np.mean(room_4[room_4['lease_commence_date']==date]['resale_price']))

plt.figure(figsize=(20,10))
plt.plot(lease_commence_date,avg_price,marker='o',color='red')
plt.title("Average Resale Price of 4 Room Flats by Lease Commence Date",fontsize=20)
plt.xticks(rotation=90)
plt.xlabel('Lease Commence Date',fontsize=15)
plt.ylabel('Average Price (SGD)',fontsize=15)
plt.grid(color="gray",alpha=0.4)
plt.show()
```

This line graph shows the relationship between the lease commence date (LCD) and the average resale price in Singapore dollars (SGD). All HDB flats are leased for a period of 99 years, and the LCD is the starting point for this calculation. A flat with a shorter lease term will be worth less than a flat with a longer lease term.

From the graph, we can see that the average resale price started with a steep increase and then decreased. The average resale price was around 600,000 SGD and it then increased to more than 900000 SGD in a few years. The price dropped back down to 450000 SGD when the LCD was 1970. Flats with an LCD between 1970 and the mid-2000s had average resale prices between 400000 SGD and 500000 SGD.

Flats with an LCD after the mid-2000s saw their average resale price increase. Although there were a few steep drops in prices, the highest it went to was above 700000 SGD. In the mid-2010s, there was a steep increase in price, from 500000 SGD to around 700000 SGD. After that steep drop, there was an increase in the average resale price, around 700000 SGD, as the LCD was closer to 2020.

Even though a flat with a recent LCD would be worth more, there are some older flats with an LCD of before 1970 which had a resale price of above 900000 SGD. There could be several factors affecting this, including the town, flat model, and facilities in that town.

As my budget is 500000 SGD, I will be choosing a 4-room flat with an LCD between 1980 and 2000.

### What is the difference between the flat models in 4 room flats?

```python
room_4 = data_resale_flat_prices[(data_resale_flat_prices['flat_type'] == '4 ROOM') & (data_resale_flat_prices['lease_commence_date'] >= np.datetime64('1980'))& (data_resale_flat_prices['lease_commence_date'] <= np.datetime64('1990'))]
flat_types = np.unique(room_4['flat_model'])
average_price=[]

for flat in flat_types:
    average_price.append(np.mean(room_4[room_4['flat_model']==flat]['resale_price']))

plt.figure(figsize=(20,10))
plt.bar(flat_types,average_price,color='red')
plt.title("Average prices of 4 room flats across different flat models",fontsize=20)
plt.xticks(rotation=90)
plt.xlabel('Flat Models',fontsize=15)
plt.ylabel('Average Resale Price (SGD)',fontsize=15)
plt.grid(color="gray",alpha=0.6)
plt.show()
```

This is a bar chart showing the different flat models and their average resale price. There are four different flat models for 4 room flats, and they are Improved, Model A, New Generation and Simplified.

Improved flats are the largest, with an average size of 105 square meters. They were built between 1982 and 1990 and have a refuse chute in the kitchen. Improved flats are the most expensive, with an average resale price of 600000SGD. Model A flats are similar in size to Improved flats, with an average size of 100 square meters. They were built between 1990 and 1998 and do not have a refuse chute in the kitchen. Model A flats are less expensive than Improved flats, with an average resale price of 450000SGD. New Generation flats are smaller than Improved and Model A flats, with an average size of 90 square meters. They were built between 1998 and 2004 and have a different layout, with the master bedroom located at the back of the flat. New Generation flats costs above 400000SGD. Simplified flats are the smallest of the four models, with an average size of 85 square meters. They were built after 2004 and have a different layout, with the master bedroom located at the front of the flat. Simplified flats are the least expensive of the four models, with an average resale price of around 300000SGD.

The most <strong>expensive</strong> flat model is the Improved flat model, which costs 600000SGD. The <strong>cheapest</strong> flat model is Simplified flat model, which costs below 400000SGD.

The average resale prices of flat models are positively correlated with the floor area of the flat model. This means that as the floor area of a flat model increases, so does its average resale price.

I will be choosing the Model A flat model because it is the most cost-effective option.

Info of flat models are from https://www.hdb.gov.sg/ .

### Do flats in higher storeys cost more?

```python
room_4 = data_resale_flat_prices[(data_resale_flat_prices['flat_type'] == '4 ROOM') & (data_resale_flat_prices['lease_commence_date'] >= np.datetime64('1980'))& (data_resale_flat_prices['lease_commence_date'] <= np.datetime64('2000')) & (data_resale_flat_prices['flat_model']=='Model A')]
storey_ranges = np.unique(room_4['storey_range'])

average_price_storey = []

for storey in storey_ranges:
    average_price_storey.append(np.mean(room_4[room_4['storey_range']==storey]['resale_price']))

plt.figure(figsize=(20,8))
plt.bar(storey_ranges,average_price_storey)
plt.title("Average prices of Model A 4 room flats across different storey ranges",fontsize=20)
plt.xticks(rotation=90)
plt.xlabel('Storey Ranges',fontsize=15)
plt.ylabel('Average Resale Price (SGD)',fontsize=15)
plt.grid(color="gray",alpha=0.6)
plt.show()
```

This is a bar chart showing the relationship between the storey range and the average resale price of Model A 4 room flats. As the storey range gets higher, the average resale price of the flat increases too. The highest storey range for a Model A 4 room flat is 28 to 30. A flat at this storey range costs more than <strong>700000SGD</strong>. The <strong>cheapest</strong> flat is also the lowest storey range, floors 1 to 3, which costs just above <strong>400000SGD</strong>.

The average resale price of the flats have been increasing as the storey range increased. However, there is an anomaly. The average resale price of a flat on storey range 19 to 21 costs close to 600000SGD. The next storey range,21 to 24 costs around 600000SGD, but it is a bit cheaper than the previous storey range. There might be several factors affecting this.

In conclusion, flats in higher storey ranges do cost more. However, there are some factors that can affect the resale price of a flat, even if it is on a higher storey range. Buyers should carefully consider all of these factors when making a decision about which flat to buy.

Based on the information in the bar chart, I would choose a Model A 4-room flat on storey range 13 to 15. This storey range has an average resale price of around 500000SGD, which is a good value for money. The flats on this storey range are also likely to have good views and plenty of natural light.


### Is there a relationship between the floor area of a property and its resale price?


```python
room_4 = data_resale_flat_prices[(data_resale_flat_prices['flat_type'] == '4 ROOM') & (data_resale_flat_prices['lease_commence_date'] >= np.datetime64('1980')) & (
    data_resale_flat_prices['lease_commence_date'] <= np.datetime64('2000')) & (data_resale_flat_prices['flat_model'] == 'Model A') & (data_resale_flat_prices['storey_range'] == '13 TO 15')]

sqm = room_4['floor_area_sqm']
resale_price = room_4['resale_price']

plt.figure(figsize=(20, 10))
a, b = np.polyfit(sqm, resale_price, 1)
plt.scatter(sqm, resale_price)
plt.plot(sqm, a*sqm+b, color='red')
plt.title("Scatter plot of floor area vs resale price for 4 Room flats", fontsize=20)
plt.xlabel('Floor Area (sqm)', fontsize=15)
plt.ylabel('Resale Price (SGD)', fontsize=15)
plt.grid(color="gray", alpha=0.4)
plt.show()
```

The scatter plot shows the relationship between the floor area and resale price of Model A 4-room flats. There is a low positive correlation between the two variables, meaning that as the floor area increases, the resale price also tends to increase. However, there are some outliers, such as the most expensive flat which has a floor area of around 115sqm but costs close to 900000 SGD.

The smallest flat has a floor area of 90 sqm and a resale price of 300000 SGD. The biggest flat has a floor area of 130 sqm but is not the most expensive flat. The most expensive flat had a floor area of around 115 sqm and costs close to 900000 SGD.Most flats have a floor area of around 100 to 110 sqm. The cheapest flat in this range costs 300000 SGD. The most expensive flat in this range costs above 800000 SGD.

In conclusion, there is a weak positive correlation between the floor area of a flat and its resale price. This means that, on average, flats with larger floor areas tend to sell for more than flats with smaller floor areas. However, there are many other factors that can affect the resale price of a flat, such as its town and facilities.

Given my budget, I will be choosing a flat with a floor area of 100 to 110 sqm. There are plenty of options available in this range.

### Which town in Singapore has Model A 4-room flats with a lease commence date between 1980 and 2000, a floor area between 100 and 110sqm, and a price around 500000 SGD?

```python
room_4 = data_resale_flat_prices[(data_resale_flat_prices['flat_type'] == '4 ROOM') & 
                                 (data_resale_flat_prices['lease_commence_date'] >= np.datetime64('1980')) & 
                                 (data_resale_flat_prices['lease_commence_date'] <= np.datetime64('2000')) &
                                 (data_resale_flat_prices['floor_area_sqm'] >= 100) &
                                 (data_resale_flat_prices['floor_area_sqm'] <= 110) & 
                                 (data_resale_flat_prices['flat_model'] == 'Model A') & 
                                 (data_resale_flat_prices['storey_range'] == '13 TO 15')]

towns = np.unique(room_4['town'])
average_price = []

for town in towns:
    average_price.append(
        np.mean(room_4[room_4['town'] == town]['resale_price']))

plt.figure(figsize=(20, 10))
plt.bar(towns, average_price)
plt.title("Average prices of 4 room flats across different towns", fontsize=20)
plt.xticks(rotation=90)
plt.xlabel('Towns', fontsize=15)
plt.ylabel('Average Resale Price (SGD)', fontsize=15)
plt.grid(color="gray", alpha=0.4)
plt.show()
```

This is a bar chart comparing the average resale price of a 4 room flat across different towns in Singapore. All flats are Model A 4 room flats, with a lease commence date between 1990 to 2000, and a floor area of 100 to 110sqm.
